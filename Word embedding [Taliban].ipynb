```python
import nltk
import urllib
import bs4 as bs
import re
from gensim.models import Word2Vec
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Taliban').read()
```

```python
source
```

```python
soup = bs.BeautifulSoup(source,'lxml')
```

```python
soup
```

```python
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
text = re.sub(r'\[[0-9]*\]',' ',text)
text = re.sub(r'\[[a-z]*\]',' ',text)
text = text.lower()
text = re.sub(r'[@#\$%\*\(\)\<\>\?\'\",:;]',' ',text)
text = re.sub(r'\xa0',' ',text)
text = re.sub(r'\u2009',' ',text)
text = re.sub(r'\n[a-z]*',' ',text)
text = re.sub(r'n/',' ',text)
text = re.sub(r'/tlbn',' ',text)
text = re.sub(r't lb',' ',text)
text = re.sub(r'\s+[a-z]\s+',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
text
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences
```

```python
len(sentences)
```

```python
sentences = [ nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sentences
```

```python
model = Word2Vec(sentences,min_count=4)
```

```python
model.wv['islamic']
```

```python
model.wv.most_similar('american')
```
